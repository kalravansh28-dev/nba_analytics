```python
# Sort the entire dataset by Team Name first, then chronologically by Game Date
games = games.sort_values(['TEAM_NAME', 'GAME_DATE']).reset_index(drop=True)
```

```python
games[games['TEAM_NAME'] == 'Atlanta Hawks'].tail(10)
```

```python
games['IS_HOME'] = games['MATCHUP'].str.contains('vs.').astype(int)
```

```python
games['OPPONENT'] = games['MATCHUP'].str.extract(r'(?:vs\.|@)\s+([A-Z]+)')
```

```python
games['GAME_NUM'] = games.groupby('TEAM_NAME').cumcount() + 1
games[games['TEAM_NAME'] == 'Atlanta Hawks'].head(10)
```

```python
window = 5

# Group by team and apply shifting rolling stats to avoid leakage
for stat in ['PTS', 'AST', 'TOV', 'REB', 'FG_PCT']:
    games[f'{stat}_ROLL{window}'] = (
        games.groupby('TEAM_NAME')[stat]
        .transform(lambda x: x.shift().rolling(window).mean())
    )
```

```python
games[['TEAM_NAME', 'GAME_DATE', 'PTS', 'PTS_ROLL5', 'AST_ROLL5', 'TOV_ROLL5', 'REB_ROLL5', 'FG_PCT_ROLL5']].head(15)
```

```python
games.columns
```

```python
# 1. Isolate the unique mappings of team IDs and abbreviations
team_id_table = games[['TEAM_ID', 'TEAM_ABBREVIATION']].drop_duplicates()

# 2. Rename columns to represent the opponent's perspective
opponent_id_table = team_id_table.rename(columns={
    'TEAM_ABBREVIATION': 'OPPONENT',
    'TEAM_ID': 'OPPONENT_TEAM_ID'
})

# 3. Merge the opponent team IDs directly back into our games dataset
games = games.merge(opponent_id_table, on='OPPONENT', how='left')
```

```python
games['home_team_game_id'] = games['GAME_ID'].astype(str) + "_" + games['TEAM_ID'].astype(str)
games['away_team_game_id'] = games['GAME_ID'].astype(str) + "_" + games['OPPONENT_TEAM_ID'].astype(str)
```

```python
opponent_stats = games[[
    'home_team_game_id', 'PTS_ROLL5', 'AST_ROLL5', 'TOV_ROLL5', 'REB_ROLL5', 'FG_PCT_ROLL5'
]].rename(columns={
    'home_team_game_id': 'away_team_game_id',
    'PTS_ROLL5': 'OPP_PTS_ROLL5',
    'AST_ROLL5': 'OPP_AST_ROLL5',
    'TOV_ROLL5': 'OPP_TOV_ROLL5',
    'REB_ROLL5': 'OPP_REB_ROLL5',
    'FG_PCT_ROLL5': 'OPP_FG_PCT_ROLL5'
})
```

```python
games = games.merge(opponent_stats, on='away_team_game_id', how='left')
```

```python
games['TARGET_WIN'] = (games['WL'] == 'W').astype(int)
```

```python
feature_columns = [
    'IS_HOME',
    'PTS_ROLL5', 'AST_ROLL5', 'TOV_ROLL5', 'REB_ROLL5', 'FG_PCT_ROLL5',
    'OPP_PTS_ROLL5', 'OPP_AST_ROLL5', 'OPP_TOV_ROLL5', 'OPP_REB_ROLL5', 'OPP_FG_PCT_ROLL5'
]
```

```python
X = games[feature_columns].dropna().copy()
y = games.loc[X.index, 'TARGET_WIN']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, shuffle=True, random_state=42
)
```

```python
scaler = StandardScaler()

# Fit scaler on training features and transform
X_train_scaled = scaler.fit_transform(X_train)

# Transform test features using the exact same scaling parameters
X_test_scaled = scaler.transform(X_test)
```

```python
print("Missing values per feature:\n", X.isnull().sum())
```

```python
print("\nTarget class distribution (1 = Win, 0 = Loss):\n", y.value_counts(normalize=True))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
```

```python
logreg = LogisticRegression(random_state=42, max_iter=1000)
logreg.fit(X_train_scaled, y_train)
```

```python
y_pred = logreg.predict(X_test_scaled)
y_prob = logreg.predict_proba(X_test_scaled)[:, 1]
```

```python
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("ROC AUC Score:", roc_auc_score(y_test, y_prob))
```

```python
def add_win_streak_feature(df):
    df = df.sort_values(['TEAM_NAME', 'GAME_DATE'])
    df['WIN_STREAK'] = 0
    
    for team in df['TEAM_NAME'].unique():
        streak = 0
        team_idx = df[df['TEAM_NAME'] == team].index
        
        for i in team_idx:
            df.at[i, 'WIN_STREAK'] = streak
            if df.at[i, 'TARGET_WIN'] == 1:
                streak += 1
            else:
                streak = 0
    return df
```

```python
def add_rest_days(df):
    df = df.sort_values(['TEAM_NAME', 'GAME_DATE'])
    df['GAME_DATE'] = pd.to_datetime(df['GAME_DATE'])
    df['REST_DAYS'] = df.groupby('TEAM_NAME')['GAME_DATE'].diff().dt.days.fillna(0).astype(int)
    return df
```

```python
def add_date_ordinal(df):
    df['DATE_ORDINAL'] = pd.to_datetime(df['GAME_DATE']).map(pd.Timestamp.toordinal)
    return df
```

```python
def add_opponent_win_rate(df):
    df = df.copy()
    df['GAME_DATE'] = pd.to_datetime(df['GAME_DATE'])
    
    df = df.sort_values(['TEAM_ID', 'GAME_DATE'])
    
    df['TEAM_WINS_CUMSUM'] = df.groupby('TEAM_ID')['TARGET_WIN'].cumsum() - df['TARGET_WIN']
    df['TEAM_GAMES_CUMSUM'] = df.groupby('TEAM_ID').cumcount()
    df['TEAM_WIN_RATE'] = df['TEAM_WINS_CUMSUM'] / df['TEAM_GAMES_CUMSUM'].replace(0, 1)
    
    winrate_lookup = df[['TEAM_ID', 'GAME_DATE', 'TEAM_WIN_RATE']].copy()
    winrate_lookup.rename(columns={
        'TEAM_ID': 'OPPONENT_TEAM_ID',
        'GAME_DATE': 'OPP_DATE',
        'TEAM_WIN_RATE': 'OPP_WIN_RATE'
    }, inplace=True)
    
    df = df.merge(
        winrate_lookup,
        how='left',
        left_on=['OPPONENT_TEAM_ID', 'GAME_DATE'],
        right_on=['OPPONENT_TEAM_ID', 'OPP_DATE']
    )
    
    df['Opponent_Win_Rate'] = df['OPP_WIN_RATE'].fillna(0.5)
    df.drop(columns=['TEAM_WINS_CUMSUM', 'TEAM_GAMES_CUMSUM', 'OPP_DATE', 'OPP_WIN_RATE'], inplace=True)
    return df
```

```python
games=add_win_streak_feature(games)
games=add_rest_days(games)
games=add_date_ordinal(games)
games=add_opponent_win_rate(games)
```
